```python
import os
import sys
from reportlab.pdfgen import canvas

def hello(c):
    c.drawString(100, 100, "Hello World")
    c.line(0, 0, 100, 100)
    
c = canvas.Canvas("hello.pdf", pagesize=(400,400))
hello(c)
c.save()
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4, letter

def hello(c):
    c.translate(A4[0]/2, A4[1]/2)
    
    c.setStrokeColorRGB(0.2, 0.5, 0.3)
    c.setFont("Helvetica", 14)
    for deg in range(0, 360, 30):
        c.drawString(50, 0, "Hello World {}".format(deg))
        c.rotate(30)
    c.setFillColorRGB(0, 0, 1)
    c.circle(0, 0, 50-3, stroke=True, fill=True)
    
c = canvas.Canvas("hello2.pdf", pagesize=A4)
hello(c)
c.save()
```

```python
from reportlab.pdfgen import canvas
from reportlab.pdfbase import pdfmetrics
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.lib.pagesizes import A4, letter

def hello(c):
    c.translate(A4[0]/2, A4[1]/2)
    
    c.setStrokeColorRGB(0.2, 0.5, 0.3)
    pdfmetrics.registerFont(TTFont("개성", "Gaesung.ttf"))
    c.setFont("개성", 14)
    for deg in range(0, 360, 30):
        c.drawString(50, 0, "한글 세종대왕")
        c.rotate(30)
    c.setFillColorRGB(0, 0, 1)
    c.circle(0, 0, 50-3, stroke=True, fill=True)
    
c = canvas.Canvas("hello3.pdf", pagesize=A4)
hello(c)
c.save()
```

```python
import glob
from reportlab.pdfgen import canvas
from reportlab.pdfbase import pdfmetrics
from reportlab.lib.pagesizes import A4
from PIL import Image

flist = glob.glob(r'C:\Users\admin\Pictures\*.png')
```

```python
c = canvas.Canvas('twice.pdf', pagesize=A4)
for fpath in flist:
    im = Image.open(fpath)
    w, h = im.size
    r = w / h
    w2 = A4[0]
    h2 = w2 / r
    c.drawImage(fpath, 0, 0, w2, h2)
    c.showPage()
c.save()
```
